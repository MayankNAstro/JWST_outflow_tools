```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
from photutils.centroids import centroid_com, centroid_2dg, centroid_1dg
from astropy.wcs import WCS
# --- Load FITS file ---
filename = "Cont14.fits"
with fits.open(filename) as hdul:
    data = hdul[0].data
    header = hdul[0].header

# If the FITS file has extra dimensions (e.g., [1, :, :]), squeeze them



# --- Compute centroids ---

dg2_centroid = centroid_2dg(data)     



print("2D Gaussian centroid   :", dg2_centroid)


# --- Optional: Plot and mark the centroid ---
plt.imshow(data, origin='lower', cmap='inferno')

plt.scatter(dg2_centroid[0], dg2_centroid[1], s=100, c='lime', marker='x', label='2DG')

plt.legend()
plt.title("Photocenter of FITS Image")
plt.show()
```

```python
wcs = WCS(header)
ra, dec = wcs.wcs_pix2world(dg2_centroid[0], dg2_centroid[1], 0)  # 0 for pixel convention

print(f"2D Gaussian centroid (world): RA={ra:.6f} deg, Dec={dec:.6f} deg")
```
